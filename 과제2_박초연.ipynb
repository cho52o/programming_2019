# Hangman 과제 가이드 프로그램

```python
#1 import
import random
```

```python
HANGMANPICS = ['''

  +---+
  |   |
      |
      |
      |
      |
=========''', '''

  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''

  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''

  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''

  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''

  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''

  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']
```

```python
words = 'ant baboon badger bat bear beaver camel cat clam cobra cougar coyote '\
'crow deer dog donkey duck eagle ferret fox frog goat goose hawk '\
'lion lizard llama mole monkey moose mouse mule newt '\
'otter owl panda parrot pigeon python rabbit ram rat raven rhino '\
'salmon seal shark sheep skunk sloth snake spider stork swan '\
'tiger toad trout turkey turtle weasel whale wolf wombat zebra '
words=words.split()
```

<br><br>
* 아래 함수는 단어를 랜덤하게 선택하여 반환한다.

```python
def getRandomWord(wordList):
    #1            #wordIndex = random.randint(0, $$$$$$$$$$$$)    $$$$에 들어갈 내용은?
    wordIndex = random.randint(0, len(words))
    return wordList[wordIndex]
```

<br><br>
* 이 함수는 사용자가 단어의 문자를 추측하기 전에 필요한 정보를 보여주기 위함이다.    

```python
def displayBoard(HANGMANPICS, missedLetters, correctLetters, secretWord):
    print(HANGMANPICS[len(missedLetters)])
    print()
    
    print('Missed letters:', end=' ')
    for letter in missedLetters:
        print(letter, end=' ')
    print()

    #1 secretWord의 문자 갯수만큼의 '_'를 blanks 라는 변수에 저장한다.
    blanks=len(secretWord)*'_'

    for i in range(len(secretWord)): # replace blanks with correctly guessed letters
        if secretWord[i] in correctLetters:
            blanks = blanks[:i] + secretWord[i] + blanks[i+1:]

    #2 for문을 이용하여 콘솔에 blanks를 인쇄한다.
    for i in blanks:
        print(i, end=' ')

    print()
```

<br><br>
* 입력한 데이터를 테스트하여 정상적인 입력 문자를 return 한다.     

```python
def getGuess(alreadyGuessed):   
    while True:
        print('Guess a letter.')
        #1 입력 데이터를 받아 변수 guess에 저장한다. 
        guess=input()
        #1 guess의 값을 소문자로 바꾸어 다시 변수 guess에 저장한다.
        guess.lower()
        #1 if 만일 입력된 데이터의 크기가 1이 아니면
        if not len(guess) == 1:
            print('Please enter a single letter.')        
        #1 elif 만일 입력된 데이터가 이미 입력된데이터(alreadyGuessed) 중에 있다면
        elif alreadyGuessed == guess:
            print('You have already guessed that letter. Choose again.')
        #1 elif 만일 입력된 데이터가 알파벳이 아니라면 
        elif not (ord(guess)>96 and ord(guess)<123):
            print('Please enter a LETTER.')
        else:        
            return guess      
```

<br><br>
* 게임을 한번 더 하기를 원하는지 여부를 return 한다.

```python
def playAgain():
    print('Do you want to play again? (yes or no)')
    return input().lower().startswith('y')
```

<br><br>
* 메인 프로그램

```python
print('H A N G M A N') 
missedLetters = ''      # 변수 초기화
correctLetters = ''     # 변수 초기화
secretWord = getRandomWord(words)  # words로 부터 랜덤하게 단어하나를 선택하여 변수 secretWord 에 저장한다. 
gameIsDone = False  # 현재 게임이 끝난는지를 알려주는 부울변수 gameIsDone을 초기화 한다


while True:
    displayBoard(HANGMANPICS, missedLetters, correctLetters, secretWord) # 콘솔에 행맨그림, 틀린입력글자, 맞은 입력글자(위치포함)를 보여준다.   
    guess = getGuess(missedLetters + correctLetters) # getGuess() 함수의 결과(새로운 알파벳 입력)를 변수 guess에 저장 

    if guess in secretWord:     # 만일 입력한 문자가 secretWord에 있으면 
        #1 correctLetters를 업데이트하고 
        correctLetters+=guess

        foundAllLetters = True   # 모두 맞추었는지를 나타내는 변수 foundAllLetters를 초기화 
        
        
        for c in secretWord:    # secretWord의 각 문자에 대하여

            #1 if 만일 그 문자가 correctLetters에 없으면,
            if not (c in correctLetters):
                #2 foundAllLetters값을 False 로 바꾸고 루프 탈출
                foundAllLetters = False
                break
        
        if foundAllLetters:  #만일 모든 문자를 맞추었으면
            print('Yes! The secret word is "' + secretWord + '"! You have won!')           
            #1 현재 게임이 종료되었음을 알리도록 gameIsDone 값을 바꾼다.
            gameIsDone = True
        
    else: # 만일 입력한 문자가 secretWord에 없으면 
        #1 missedLetters를 업데이트하고,

        missedLetters+=guess
         
        if len(missedLetters) == len(HANGMANPICS) - 1: # 잘못입력한 문자의 갯수가 허용된를 넘으면 
            displayBoard(HANGMANPICS, missedLetters, correctLetters, secretWord) # 콘솔에 정보를 보여주고,
            print('You have run out of guesses!\nAfter ' + str(len(missedLetters)) + ' missed guesses and ' + str(len(correctLetters)) + ' correct guesses, the word was "' + secretWord + '"')
            #1 현재 게임이 끝났음 나타내도록 변수 gameIsDone 업데이트
            gameIsDone = True

    
    if gameIsDone:  # 만일 현재 게임(gameIsDone)이 끝났으면
        #1 if 만일 게임을 더 하기를 원하면 (playAgain() 함수 사용)
        if playAgain():
            #1 missedLetters 초기화 
            missedLetters = ''
            #1 correctLetters 초기화 
            correctLetters = ''
            #1 현재 게임(gameIsDone)이 끝나지 않은 것으로 변경
            secretWord = getRandomWord(words)
            #1 words로 부터 랜덤하게 단어하나를 선택하여 변수 secretWord 에 저장한다. 
            gameIsDone = False
            
        #1 else
        else:
            #1 루프 탈출
            break
```
